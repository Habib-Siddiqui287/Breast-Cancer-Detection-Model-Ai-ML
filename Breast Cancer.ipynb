```python
##Name Habib Ahmad Siddiqui
## Roll no 71
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score,confusion_matrix,precision_score,recall_score,f1_score)
from sklearn.preprocessing import StandardScaler,LabelEncoder
from scipy import stats
```

```python
##Name Habib Ahmad Siddiqui
## Roll no 71
path = r"C:\\Users\\Habbib Ahmed Sidique\\OneDrive\\Desktop\\AI ML\\Classifications Models\\Logistic Regression\\Breast Cancer Disease Detector Model\\archive\breast-cancer.csv"
Breast_cancer = pd.read_csv(path)
Breast_cancer.head(5)
```

```python
##Name Habib Ahmad Siddiqui
## Roll no 71
df=pd.DataFrame(Breast_cancer)
```

```python
##Name Habib Ahmad Siddiqui
## Roll no 71
Breast_cancer.head()
Breast_cancer.tail()
Breast_cancer.shape
Breast_cancer.info()
Breast_cancer.isnull().sum()
Breast_cancer.describe()
```

```python
##Name Habib Ahmad Siddiqui
## Roll no 71
encoder=LabelEncoder()
```

```python
##Name Habib Ahmad Siddiqui
## Roll no 71
for col in df.columns:
    if df[col].dtype=='object':
        df[col]=encoder.fit_transform(df[col])
         
```

```python
##Name Habib Ahmad Siddiqui
## Roll no 71
X=df.drop(columns =  'diagnosis')
Y=df['diagnosis']
print(X)
print(Y)
```

```python
##Name Habib Ahmad Siddiqui
## Roll no 71
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
print(X.shape,X_train.shape,X_test.shape)
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
##Name Habib Ahmad Siddiqui
## Roll no 71
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print('Accuracy on Training data:',training_data_accuracy)
```

```python
##Name Habib Ahmad Siddiqui
## Roll no 71
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print('Accuracy on Test Data:',test_data_accuracy)
```

```python
##Name Habib Ahmad Siddiqui
## Roll no 71
cm = confusion_matrix(Y_test,X_test_prediction)
print("confusion_matrix")
print(cm)
```

```python
##Name Habib Ahmad Siddiqui
## Roll no 71
recall = recall_score(Y_test,X_test_prediction)
print("Recall Score is: ")
print(recall)
```

```python
##Name Habib Ahmad Siddiqui
## Roll no 71
F1_score = f1_score(Y_test,X_test_prediction)
print("F1-Score is: ")
print(F1_score)
```

```python
##Name Habib Ahmad Siddiqui
## Roll no 71
precise= precision_score(Y_test,X_test_prediction)
print("Precision Score is: ")
print(precise)
```
